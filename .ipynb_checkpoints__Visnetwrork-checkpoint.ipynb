```python
# ================================================================
# Grafo dirigido (small -> big) SOLO capas 1 y 2 desde TSV (PyVis en Jupyter)
# Lee: /home/nacho/HDD16/Nacho/RepliCOOC/results/mr_vs_familyspace/edges_containment.tsv
# Requisitos: pip install pandas pyvis
# ================================================================
import math
import pandas as pd
from pathlib import Path
from pyvis.network import Network
from IPython.display import IFrame, display

# ------------------- Parámetros -------------------
TSV_PATH = "/home/nacho/HDD16/Nacho/RepliCOOC/results/mr_vs_familyspace/edges_containment.tsv"
OUT_HTML = "small_to_big_L1_L2.html"
HEIGHT   = "800px"
WIDTH    = "100%"

# ------------------- Carga y limpieza -------------------
p = Path(TSV_PATH)
if not p.exists():
    raise FileNotFoundError(f"No existe el archivo: {p}")

df = pd.read_csv(p, sep="\t", dtype=str)

needed = ["small_id","small_combo","small_n_replicons",
          "big_id","big_combo","big_n_replicons",
          "fraction_present","n_present","n_genes","candidate_type"]
missing = [c for c in needed if c not in df.columns]
if missing:
    raise SystemExit(f"Faltan columnas: {missing}")

# Cast numéricos
for col in ["small_n_replicons","big_n_replicons","fraction_present","n_present","n_genes"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Deduplicar por (small_id, big_id) conservando mayor fraction_present
df = (df.sort_values("fraction_present", ascending=False, na_position="last")
        .drop_duplicates(subset=["small_id","big_id"], keep="first")
        .reset_index(drop=True))

# ------------------- Filtra SOLO capas 1 y 2 -------------------
df12 = df[
    df["small_n_replicons"].isin([1,2]) &
    df["big_n_replicons"].isin([1,2])
].copy()

# Asegura dirección small->big con small capa < big capa
df12 = df12[df12["small_n_replicons"] < df12["big_n_replicons"]].copy()

if df12.empty:
    raise SystemExit("No hay aristas entre capas 1 → 2 en el TSV.")

print(f"Filtrado L1→L2: {df12['small_id'].nunique()+df12['big_id'].nunique()} nodos, {len(df12)} aristas")

# ------------------- Utilidades de estilo -------------------
def edge_width(frac, min_w=1.5, max_w=10.0):
    if pd.isna(frac): return min_w
    f = float(max(0.0, min(1.0, frac)))
    return min_w + (max_w - min_w) * f

def rgba(gray, alpha):
    g = max(0, min(255, int(gray)))
    a = max(0.05, min(1.0, float(alpha)))
    return f"rgba({g},{g},{g},{a})"

# ------------------- Construir red PyVis -------------------
net = Network(height=HEIGHT, width=WIDTH, directed=True, notebook=True, cdn_resources="remote")
net.barnes_hut(gravity=-2500, central_gravity=0.25, spring_length=160, spring_strength=0.01)

LAYER_STYLE = {
    1: dict(shape="dot",    size=18, color="#6e9cff", label_suffix=" (L1)"),
    2: dict(shape="square", size=22, color="#ffb86e", label_suffix=" (L2)"),
}

# Nodos: los que aparecen en las aristas filtradas
nodes_small = (df12[["small_id","small_combo","small_n_replicons"]]
               .drop_duplicates()
               .rename(columns={"small_id":"id","small_combo":"combo","small_n_replicons":"layer"}))
nodes_big   = (df12[["big_id","big_combo","big_n_replicons"]]
               .drop_duplicates()
               .rename(columns={"big_id":"id","big_combo":"combo","big_n_replicons":"layer"}))
nodes = pd.concat([nodes_small, nodes_big], ignore_index=True).drop_duplicates(subset=["id"])

for _, r in nodes.iterrows():
    nid   = r["id"]
    combo = r.get("combo","") or ""
    layer = int(r.get("layer", 0) or 0)
    st = LAYER_STYLE.get(layer, dict(shape="dot", size=16, color="#c7c7c7", label_suffix=""))
    title = (
        f"<b>{nid}</b>"
        f"<br>layer={layer}"
        f"<br>combo={combo.replace('|',' | ')}"
    )
    net.add_node(
        nid,
        label=nid + st.get("label_suffix",""),
        title=title,
        shape=st["shape"],
        size=st["size"],
        color=st["color"],
        physics=True
    )

# Aristas: solo L1 → L2
for _, r in df12.iterrows():
    sid, bid = r["small_id"], r["big_id"]
    frac = r["fraction_present"]
    title = (
        f"<b>{sid}</b> → <b>{bid}</b>"
        f"<br>fraction_present={('%.3f' % float(frac)) if pd.notna(frac) else 'NA'}"
        f"<br>n_present={int(r['n_present']) if pd.notna(r['n_present']) else 'NA'}"
        f"<br>n_genes={int(r['n_genes']) if pd.notna(r['n_genes']) else 'NA'}"
        f"<br>candidate_type={r.get('candidate_type','')}"
    )
    width  = edge_width(frac)
    stroke = rgba(60, 0.25 + 0.7 * (0 if pd.isna(frac) else float(frac)))
    net.add_edge(sid, bid, title=title, width=width, color=stroke, arrows="to", smooth=True)

# Opciones UI (JSON puro, no JS)
net.set_options("""
{
  "layout": { "improvedLayout": true },
  "interaction": { "hover": true, "tooltipDelay": 120, "navigationButtons": true },
  "edges": { "smooth": { "type": "dynamic", "roundness": 0.35 } },
  "nodes": { "borderWidth": 1, "shadow": true }
}
""")

# Render inline
net.show(OUT_HTML)
display(IFrame(src=OUT_HTML, width="100%", height=int(''.join(ch for ch in HEIGHT if ch.isdigit()) or 800)))
```

```python
# ================================================================
# 4-column fixed layout (one column per layer: L1, L2, L3, L4)
# - Reads edges_containment.tsv
# - Places ALL L1 nodes in column 1, L2 in column 2, etc.
# - No subcolumns, no physics. Colors by replicon (bg=1st token, border=2nd).
# - Saves an interactive HTML (does NOT display it in the notebook)
# Reqs: pip install pandas pyvis
# ================================================================
import pandas as pd
from pathlib import Path
from pyvis.network import Network

# ------------------- Params -------------------
TSV_PATH   = Path("/home/nacho/HDD16/Nacho/RepliCOOC/results/mr_vs_familyspace/edges_containment.tsv")
OUT_HTML   = TSV_PATH.with_name("edges_containment_L1toL4_4columns.html")

# Canvas & column geometry
WIDTH_PX   = 2400           # canvas width
HEIGHT_PX  = 30000          # canvas height
LEFT_MRG   = 120            # left margin
RIGHT_MRG  = 120            # right margin
TOP_MRG    = 80             # top margin
BOT_MRG    = 80             # bottom margin

# Optional family filter (set to None to include all)
REPL_PREFIX = "IncQ"          # e.g., "IncF", "IncQ", or None

# ------------------- Load -------------------
if not TSV_PATH.exists():
    raise FileNotFoundError(f"No existe el archivo: {TSV_PATH}")

df = pd.read_csv(TSV_PATH, sep="\t", dtype=str)

needed = ["small_id","small_combo","small_n_replicons",
          "big_id","big_combo","big_n_replicons",
          "fraction_present","n_present","n_genes","candidate_type"]
missing = [c for c in needed if c not in df.columns]
if missing:
    raise SystemExit(f"Faltan columnas: {missing}")

# Numerics
for col in ["small_n_replicons","big_n_replicons","fraction_present","n_present","n_genes"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Keep only edges with layers in 1..4 and small<big
df = df[
    df["small_n_replicons"].between(1,4) &
    df["big_n_replicons"].between(1,4) &
    (df["small_n_replicons"] < df["big_n_replicons"])
].copy()

# Optional: filter by family prefix in either endpoint
def has_prefix(combo: str, prefix: str) -> bool:
    if not isinstance(combo, str) or combo.strip() == "":
        return False
    return any(tok.strip().startswith(prefix) for tok in combo.split("|"))

if REPL_PREFIX:
    df = df[
        df["small_combo"].apply(lambda s: has_prefix(s, REPL_PREFIX)) |
        df["big_combo"].apply(lambda s: has_prefix(s, REPL_PREFIX))
    ].copy()

# Deduplicate small->big by highest fraction
df = (df.sort_values("fraction_present", ascending=False, na_position="last")
        .drop_duplicates(subset=["small_id","big_id"], keep="first")
        .reset_index(drop=True))

if df.empty:
    raise SystemExit("No hay aristas válidas tras el filtrado.")

# ------------------- Build node table -------------------
def tokens(combo: str):
    if not isinstance(combo, str) or combo.strip() == "":
        return []
    return [t.strip() for t in combo.split("|") if t.strip()]

nodes = pd.concat([
    df[["small_id","small_combo","small_n_replicons"]].rename(columns={
        "small_id":"id","small_combo":"combo","small_n_replicons":"layer"}),
    df[["big_id","big_combo","big_n_replicons"]].rename(columns={
        "big_id":"id","big_combo":"combo","big_n_replicons":"layer"})
], ignore_index=True).drop_duplicates(subset=["id"])

# Keep only 1..4
nodes = nodes[nodes["layer"].between(1,4)].copy()

# Colors
nodes["repA"] = nodes["combo"].apply(lambda c: tokens(c)[0] if tokens(c) else "NA")
def repB_of(c, a):
    t = tokens(c)
    return t[1] if len(t) >= 2 else a
nodes["repB"] = [repB_of(c, a) for c, a in zip(nodes["combo"], nodes["repA"])]

# Sort inside each layer by replicon then id (for stable vertical order)
nodes = nodes.sort_values(["layer","repA","id"]).reset_index(drop=True)

PALETTE = [
    "#4e79a7", "#f28e2b", "#e15759", "#76b7b2", "#59a14f",
    "#edc948", "#b07aa1", "#ff9da7", "#9c755f", "#bab0ab"
]
reps_seq = pd.unique(pd.concat([nodes["repA"], nodes["repB"]], ignore_index=True))
color_map = {r: PALETTE[i % len(PALETTE)] for i, r in enumerate(reps_seq)}

# ------------------- Exact 4 columns: compute fixed (x,y) -------------------
# Column centers evenly spaced across width
usable_w = WIDTH_PX - LEFT_MRG - RIGHT_MRG
col_sep  = usable_w / 3.0
x_by_layer = {
    1: int(LEFT_MRG + 0*col_sep),
    2: int(LEFT_MRG + 1*col_sep),
    3: int(LEFT_MRG + 2*col_sep),
    4: int(LEFT_MRG + 3*col_sep),
}

# For each layer, distribute nodes vertically from TOP to BOTTOM with equal spacing
pos_xy = {}
for L in (1,2,3,4):
    ids_L = nodes.loc[nodes["layer"]==L, "id"].tolist()
    nL = len(ids_L)
    if nL == 0:
        continue
    y_top = TOP_MRG
    y_bot = HEIGHT_PX - BOT_MRG
    if nL == 1:
        ys = [ (y_top + y_bot)//2 ]
    else:
        step = (y_bot - y_top) / (nL - 1)
        ys = [ int(y_top + i*step) for i in range(nL) ]
    xs = [ x_by_layer[L] ] * nL
    for nid, x, y in zip(ids_L, xs, ys):
        pos_xy[nid] = (x, y)

# ------------------- Build PyVis network -------------------
def edge_width(frac, min_w=1.5, max_w=10.0):
    if pd.isna(frac): return min_w
    f = float(max(0.0, min(1.0, frac)))
    return min_w + (max_w - min_w) * f

net = Network(height=f"{HEIGHT_PX}px", width=f"{WIDTH_PX}px", directed=True, notebook=False, cdn_resources="remote")

SHAPE = {1:"dot", 2:"square", 3:"diamond", 4:"triangle"}

for _, r in nodes.iterrows():
    nid   = r["id"]
    combo = r["combo"] or ""
    L     = int(r["layer"])
    repA  = r["repA"]; repB = r["repB"]
    if nid not in pos_xy:   # could happen if layer had no nodes after filters
        continue
    x, y  = pos_xy[nid]
    net.add_node(
        nid,
        label=f"{nid} (L{L})",
        title=f"<b>{nid}</b><br>layer={L}<br>combo={combo.replace('|',' | ')}",
        x=x, y=y, fixed=True, physics=False,
        shape=SHAPE.get(L, "dot"),
        color={"background": color_map.get(repA, "#c7c7c7"),
               "border": color_map.get(repB, "#888888")},
        size=20 if L==1 else (22 if L==2 else 24)
    )

# Edges small->big (already filtered small<big)
for _, r in df.iterrows():
    sid, bid = r["small_id"], r["big_id"]
    if sid not in pos_xy or bid not in pos_xy:
        continue
    frac = r["fraction_present"]
    title = (
        f"<b>{sid}</b> → <b>{bid}</b>"
        f"<br>fraction_present={('%.3f' % float(frac)) if pd.notna(frac) else 'NA'}"
        f"<br>n_present={int(r['n_present']) if pd.notna(r['n_present']) else 'NA'}"
        f"<br>n_genes={int(r['n_genes']) if pd.notna(r['n_genes']) else 'NA'}"
        f"<br>candidate_type={r.get('candidate_type','')}"
    )
    net.add_edge(sid, bid, title=title, width=edge_width(frac), arrows="to", physics=False, smooth=False)

# No physics, no auto-layout
net.set_options("""
{
  "physics": { "enabled": false },
  "interaction": { "hover": true, "tooltipDelay": 120, "navigationButtons": true },
  "edges": { "smooth": false },
  "nodes": { "borderWidth": 2, "shadow": true }
}
""")

# ------------------- Save HTML -------------------
net.save_graph(str(OUT_HTML))
print(f"✅ HTML escrito en: {OUT_HTML}")
print(f"   Nodos: {nodes.shape[0]} | Aristas: {df.shape[0]}")
print(f"   Column X positions: {x_by_layer}")
print(f"   Canvas: {WIDTH_PX}x{HEIGHT_PX}px  (ajusta TOP/BOT margins si necesitas más alto)")
```
